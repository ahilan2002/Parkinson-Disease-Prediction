### 5) Find the outliers and replace the outliers

```python
df.boxplot(column = 'age')
plt.show()
```

```python
#Handling the Outliers of Age
age_median = df['age'].median()

# Set the threshold for outliers
age_upper_threshold = df['age'].quantile(0.75) + 1.5 * (df['age'].quantile(0.75) - df['age'].quantile(0.25))
age_lower_threshold = df['age'].quantile(0.25) - 1.5 * (df['age'].quantile(0.75) - df['age'].quantile(0.25))

# Replace outliers in 'Age' with the median
df.loc[df['age'] > age_upper_threshold, 'age'] = age_median
df.loc[df['age'] < age_lower_threshold, 'age'] = age_median
```
